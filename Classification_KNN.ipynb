#KNN implementation

```python
import numpy as np#
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('https://raw.githubusercontent.com/Abeyaantrix2Internship/intern/main/Social_Network_Ads.csv') 
Data.head()
#age and estimated salary columns are features(predictors)
# purchase columns is target
```

```python
Data.shape
```

```python
Data.describe()
```

```python
Data.describe(include='all')
```

```python
#split data into predictors and targets
Predictors= Data.iloc[:,2:4]
Target= Data.iloc[:,-1:]
print(Predictors.head())
print(Target.head())
```

```python
# apply standardization inorder to bring column into same scale
# standardization data is vaey important in classification
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
X=SS.fit_transform(Predictors)
```

```python
print(X)
```

```python
#feature engineering
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X, Target,test_size=0.2,random_state=0)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
#KNN classifier 
from sklearn.neighbors import KNeighborsClassifier
KNN= KNeighborsClassifier(n_neighbors=3, metric='minkowski',p=2) #When p=2 minikowski becomes euclidian distance
KNN.fit(X_train,Y_train)
```

```python
#now check test data
Y_predicted=KNN.predict(X_test)#Y predicted data from X_test data
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
CM=confusion_matrix(Y_test,Y_predicted)
print(CM)
```

```python
Accuracy=((CM[0,0]+CM[1,1])/sum(sum(CM)))*100
print('accuracy of model=',Accuracy,"%")
```

```python
Precision= (CM[0,0]/(CM[0,0]+CM[1,0]))*100
print('precision of model=',Precision,'%')
```

```python
Recall=(CM[0,0]/(CM[0,0]+CM[0,1]))*100
print('Recall of model=',Recall,'%')
```

```python
print('F-Score of model=',(2*Precision*Recall/(Precision+Recall)),'%')
```

```python
from sklearn.metrics import precision_recall_fscore_support
P,RC,F1,Array= precision_recall_fscore_support(Y_test, Y_predicted)
```

```python
print("precesion=",P[0],'\n','Recall=',RC[0],'\n','F-Score=',F1[0])
```
