```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data=pd.read_csv('https://raw.githubusercontent.com/Abeyaantrix2Internship/intern/main/FuelConsumption.csv')#
Data.head()
```

```python
#EDA
#split data into dependent and independent variables
Y= Data.iloc[:,-1:]# dapendent variables
X= Data.iloc[:,[4,5,7]]# independent variables
```

```python
print(X.FUELTYPE.unique())
print('No of columns in Fuel type Column:',len(X.FUELTYPE.unique()))
```

```python
print(X.shape)
print(Y.shape)
```

```python
# now apply Data preprocessing on fuel type columns
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
CT= ColumnTransformer([('encoder',OneHotEncoder(),[2])], remainder='passthrough')
#X_En=CT.fit_transform(X)
X_En=np.array(CT.fit_transform(X), dtype=np.float64)
print(X_En)
```

```python
#removing Dummy column
X_encoded=X_En[:,1:]
print(X_encoded)
```

```python
## Feature engg
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X_encoded,Y, test_size=0.2, random_state=0)
```

```python
# apply linear regression
from sklearn.linear_model import LinearRegression
MLR= LinearRegression()
MLR.fit(X_train,Y_train)
```

```python
#check on test data
Y_predicted= MLR.predict(X_test)
```

```python
MLR.coef_
```

```python
MLR.intercept_
```

```python
#we need to clearly identify which are the variables having influence in predicting the outcome
# here with increase in 1 unit of ingine size increases co2 emission by 29 times
# ecxh unit increases in cylinder will increase co2 emission by 8 times
```

```python
from sklearn.metrics import mean_squared_error
MSE= mean_squared_error(Y_predicted, Y_test)
print(MSE)
print(np.sqrt(MSE))
```

```python
from sklearn.metrics import r2_score
R2= r2_score(Y_predicted,Y_test)
print(R2)
#here we increased no of predictor variables , which yeilded improvement in results compared to using single variable
#R2 increased from 0.62 to 0.65
```

```python
import statsmodels.api as ols
X0=np.ones((853,1))# created array same size as training data(853)
X1=X_train
X_OLS=np.append(arr=X0, values=X1, axis=1)#append training data with X0
print(X_OLS.shape)
```

```python
MLR_OLS = ols.OLS(Y_train, X_OLS).fit()# summary of training data
print(MLR_OLS.summary())
```

```python
#significance value is the acceptance level
# in case of MLR ,p value will tell how significant a variable is in predicting the outcome
#we can remove the variable which has lower significance levels
''' 
   constant p=0.00 indicates highly significant variable
   x1       p=0.171  17% insignificance variable
   x2       p=0.925  92.5% highly insignificant variable
   x3       p=0.618  61.8% highly insignificance variable
   x4       p=0.00   0%  highly significance variable
   x5       p=0.00   0%  highly significance variable'''

# from this it is evident that fuel type (x1,x2,x3) doesnot contribute much in predicting co2 emission
```

** Durbin Watson **
- The Durbin Watson test is used to used to check the autocorrelation b/w residual
- if the value 
  
  DW close to 2 we can say no autocorrelation 
  
  if DW between 0-2 : positive Autocorelation

  if DW between 2-4 : Negative autocorelation

- in the statstical summary value is 2.08 which is close to 2 we can no autocorelation

** Jarque-Bera JB **
- it is used to check normality of the residuals JB< 5 we can say distribution is not normally otherwise normally distributed

** Condition NO **
- it is used for checking multicollinearity
- we should avoid multicollinearity within the data

1. CN<: NO MC
2. CN>100 and < 1000 :Moderate MC
3. CN> 1000: Severe MC

* from our stast summary CN= 86 which shows there is no Multicollinearty

** Optimisation of regression model **
- Backward selection method 
- forward selection method
- forward Backward selection method
- grid serch CV

** Backward selection method** 
- first step use all variables and build Regression model
- set the p value (significance level) (std value is 5%, 0.05),
- remove the highly insignificant variable
- build model once again and repeat steps 1-4 till there are no significant values

```python
import statsmodels.api as ols
X_BE = X_OLS[:,[0,1,2,3,4,5]]# x0,x1,x2,x3,x4,x5 --> Y=x0b0+x1b1+x2b2+.....x5b5
MLR_OLS = ols.OLS(Y_train, X_BE).fit()# summary of training data
print(MLR_OLS.summary())
```

```python
# x0=0.0, x1=0.17, x3=0.92(High), x3=0.61, x4=0, x5=0
#according to statistic remove X2 and build the regression model
#remove X2
X_BE = X_OLS[:,[0,1,3,4,5]]# x0,x1,x2,x3,x4,x5 --> Y=x0b0+x1b1+x2b2+.....x5b5

#x0=x0, x1=x1, x2=x3, x3=x4, x4=x5
MLR_OLS = ols.OLS(Y_train, X_BE).fit()# summary of training data
print(MLR_OLS.summary())
```

```python
# x0=0.0, x1=0.018, x2=0.26(High), x3=0.0, x4=0
#according to statistic remove X2 and build the regression model
#remove X2
X_BE = X_OLS[:,[0,1,4,5]]# x0,x1,x2,x3,x4,x5 --> Y=x0b0+x1b1+x2b2+.....x5b5

#x0=x0, x1=x1, x2=x3, x3=x4, x4=x5
#x0=x0, x1=x1, x2(x3)=x4, x3(x4)=x5
MLR_OLS = ols.OLS(Y_train, X_BE).fit()# summary of training data
print(MLR_OLS.summary())
```
